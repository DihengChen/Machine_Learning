```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import pydotplus
```

#### Read the data

```python
with open('Wisconsin_Breast_Cancer_Data.txt') as f:
    data=f.readlines()
    dataset=[]
    for lines in data:
        dataset.append(lines.rstrip('\n').split(','))
```

• Id - subject Id  
• V1 to V9 - attributes  
• Class - diagnosis: 4 = malignant, 2 = benign

```python
datBreastCancer=pd.DataFrame(dataset,columns=['ID','V1','V2','V3','V4','V5','V6','V7','V8','V9','class'])
```

```python
#change datatype to float and replace '?' by '999'
sum(np.sum(datBreastCancer=='?'))
```

```python
datBreastCancer=datBreastCancer.replace('?',999)
```

```python
datBreastCancer=datBreastCancer.astype('float')
```

```python
datBreastCancer.head()
```

#### Fit classification tree

```python
X=datBreastCancer.iloc[:,1:10]
Y=datBreastCancer.iloc[:,[-1]]
```

```python
clf = DecisionTreeClassifier()
clf.fit(X, Y)
```

```python
dot_data=tree.export_graphviz(clf, out_file=None,feature_names=datBreastCancer.columns[1:10],\
                    class_names=['2','4'],filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)  
```

```python
from IPython.display import Image
Image(graph.create_png())
```

#### Prune the tree to make analysis easier

```python
clf = DecisionTreeClassifier(max_depth=4,min_samples_leaf=10)
clf.fit(X, Y)
```

```python
dot_data=tree.export_graphviz(clf, out_file=None,feature_names=datBreastCancer.columns[1:10],\
                    class_names=['2','4'],filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())
```
